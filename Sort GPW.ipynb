```
print('Sort Population Denisty from CIESIN GPW and Transform')
```

```
%sql
DROP TABLE gpw_v4_pop_density_non_zero;
CREATE TABLE gpw_v4_pop_density_non_zero
    AS select * from gpw_v4_population_density_rev11_2020_30_sec where _c2 > 0;
CACHE TABLE gpw_v4_pop_density_non_zero
```

```
%sql
select count(0) from gpw_v4_pop_density_non_zero
```

```
%sql
select max(_c2) from gpw_v4_pop_density_non_zero
```

```
%sql
select percentile(_c2, 0.9999) from gpw_v4_pop_density_non_zero
```

```
%sql
select percentile_approx(_c2, 0.9999,10000) from gpw_v4_pop_density_non_zero
```

```
gpw = spark.sql("select * from gpw_v4_pop_density_non_zero")
```

```
gpw_sorted = gpw.sort('_c2',ascending=False)
display(gpw_sorted)
```
